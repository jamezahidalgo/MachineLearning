#**Pandas (Data Frames)**

El tipo de datos que más frecuentemente vamos a utilizar en la materia es el dataframes de Pandas. Un dataframe esencialmente es una tabla con columnas y filas indexadas, de forma tal que podemos acceder fácilmente a filas mediante el índice. También podemos unir, separar o filtrar dataframes trabajando sobre el índice. En general, los dataframes le van a resultar bastante familiares a quienes sepan SQL.

Los dataframes pueden tener datos no heterogéneos y además son objetos con métodos útiles para extraer información de forma rápida (por ejemplo, computar histogramas de valores). También es posible convertirlos a arrays de Numpy si así lo necesitamos, aunque no lo vamos a hacer a menos que sea realmente necesario.

Para usar la libreria de Pandas, debemos importarla. Comunmente lo haremos así:

````python
import pandas as pd
````

donde "as pd" es darle un pseudonimo a pandas para escrbirlo más corto. Normalmente lo encontrará así en Internet ;)

Una forma de construir dataframes es a partir de diccionarios, de forma tal que asignamos una lista a cada llave, que representa una columna del dataframe:

````python
datos = {'alumno': ['Zutano', 'Mengano', 'Pepe' ,'Fulanito, Cosme', 'Maria'],
         'primer parcial': [7, 8,7,4, 10], 'segundo parcial': [10,9,4,10,10],
         'observaciones':['ninguna','libre','ninguna','libre','oyente'],
         'DNI':[23000000, 12389100, 99999, 1001,30406011]}

df = pd.DataFrame(datos)
print(df)
````

Vemos que las filas están identificadas con números (0,1,2,...). Esos números corresponden al índice (index).

Podemos ademas seleccionar una columna del dataframe como si se tratase de un diccionario. El objeto resultante se llama "series":

````python
print(type(df))
print(df['alumno'])
print(type(df['alumno']))
````

Para obtener filas en vez de columnas, podemos usar el método df.loc con una lista de los índices que queremos acceder:

````python
df.loc[[2,3,4]]
````

Algunas de los métodos que podemos aplicar a los dataframe son útiles para tener una idea de los valores de una columna sin tener que hacer demasiado trabajo. Esto es lo que podemos hacer con value_counts(), que nos devuelve todos los valores que toman las entradas de la columna, y la cantidad de veces que ocurren.

````python
df['observaciones'].value_counts()
````

#**Matplotlib**

[Matplotlib](https://matplotlib.org/) es un paquete de python que permite realizar graficos 2d, y 3d, entre otras cosas.

Para usar la libreria, primero la importamos

````python
# Declaracion permite que el grafico generado aparesca en el notebooks
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
matplotlib.axes.Axes.boxplot
matplotlib.pyplot.boxplot
````

Creamos un primer gráfico

````python
# Creamos un objeto iterable (puede ser una lista o un arreglo) con 11 elementos empezando en -10 o sea que hacemos del -10 al 10 para x.
x = range(-10, 11)

# Creamos una lista llamada "y" con los cuadrados de los valores de x.
y = [z**2 for z in x]

# Definimos que vamos a graficar.
plt.plot(x, y)

# Definimos los labels del eje x,y.
plt.xlabel('Tiempo')
plt.ylabel('Cantidades')

# Generamos el grafico.
plt.show()
````

#Creamos un gráfico de dispersión

Estos gráficos sirven para revisar comportamientos que tienen 2 variables.

````python
# Importamos numpy para crear los datos. Esto no lo habiamos hecho antes, pero es recomendable hacerlo junto con importar Pandas al comienzo.
import numpy as np

# Fijamos la semilla. La semilla es un número que tiene las funciones que generan números aleatorios
np.random.seed(19)

# generamos números aleatorios segun una formula matemática.
x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0

plt.scatter(x, y, c="g", alpha=0.5, label="descripcion del dato")
plt.xlabel("variable X")
plt.ylabel("variable y")
plt.legend(loc='upper left')
plt.show()
````

# Creamos un gráfico de boxplot (o caja)

Este gráfico sirve para visualizar la distribución de los puntos de una variable en particular. Puede entregar mucha información.

````python
# Primero se creara datos de manera aleatoria. Variable aleatoria uniforme!!
# Se crean datos para que existan tambien datos en los extremos (outliers)
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
data = np.concatenate((spread, center, flier_high, flier_low))
````

````python
fig1, ax1 = plt.subplots()
ax1.set_title('Gráfico básico', fontsize=14, fontweight="bold")
ax1.boxplot(data)
plt.show()
````

Este gráfico entrega los cuartiles 1, 2 y 3. Tambien llamados percentil 25, 50 y 75.

La linea horizontal roja muestra el punto donde esta el cuartil 2 (Q2). Luego la linea inferior y superior de la caja (cuadrado) son el Q1 y Q3 respectivamente. La diferencia de estos (Q3 - Q1) se le llama rango intercuartil o IQR en ingles.

A continuación, una imágen que explica lo demas:


boxplot-histograma.png



````python
# Ahora un boxplot horizontal
red_square = dict(markerfacecolor='r', marker='s')
fig2, ax2 = plt.subplots()
ax2.set_title('Lo mismo pero con una pequeña mejora gráfica')
ax2.boxplot(data, flierprops=red_square, vert=False, whis=0.75)
plt.show()
````

Por último, podemos hacer varios boxplot en un mismo gráfico. La idea es comparar distribuciones de datos de distintas variables.

````python
spread = np.random.rand(50) * 100
center = np.ones(50) * 40  # Ajustar la longitud de center para que coincida con la longitud de spread
flier_high = np.random.rand(50) * 100 + 100  # Ajustar la longitud de flier_high
flier_low = np.random.rand(50) * -100  # Ajustar la longitud de flier_low

# Crear un array 2D concatenando los datos
data = np.column_stack([spread, center, flier_high, flier_low])

fig7, ax7 = plt.subplots()
ax7.set_title('Multiple Samples with Different sizes')
ax7.boxplot(data)

plt.show()

````

### Graficamos Superficies en 3D.



````python
# Importamos Axes3D lo que nos va a permitir graficar en 3D
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Creamos los datos a graficar (funciones matemátcas. Solo datos de ejemplo)
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)   # aca se crea una grilla de puntos para posteriormente graficar el punto en la tercera coordenada
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Preparamos Figure para graficar la superficie.
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definimos graficar X, Y y Z como superficie.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

#Agregamos referencia grafica de los valores (Opcional).
fig.colorbar(surf, shrink=0.5, aspect=5)

# Generamos el grafico.
plt.show()
````

### Multiples gráficos juntos

En este caso vamos a realizar 4 graficos simples en el mismo plot.

````python
# Generamos los datos de prueba.
np.random.seed(1968)             # esto es para generar numeros aleatoreos
data = np.random.randn(2, 100)

# Creamos un subplot de 2 columnas con 2 filas cada una
fig, axs = plt.subplots(2, 2, figsize=(5, 5))

# Agregamos un gráfico de histograma en la posicion 0, 0
axs[0, 0].hist(data[0])

# Agregamos un gráfico de puntos en la posicion 1, 0
axs[1, 0].scatter(data[0], data[1])

# Agregamos un gráfico de lineas en la posicion 0, 1
axs[0, 1].plot(data[0], data[1])

# Agregamos un gráfico de histograma 2d en la posición 1, 1
axs[1, 1].hist2d(data[0], data[1])

# Generamos el grafico.
plt.show()
````
